## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
listings.dtypes
```

```python
# 1. How many listings are there with a price less than 100? 
cheap_listings = int((listings["price"] < 100).sum())
print(f'There are {cheap_listings} listings under $100')


```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.
df_bk = listings[listings["neighbourhood_group"] == "Brooklyn"].copy()
df_bk.shape[0]

```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.
cheap_bk_listings = (df_bk["price"] < 100).sum()
print(f'There are {cheap_bk_listings} listings under $100 in bk')

```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names
hosts = ["Michael", "David", "John", "Daniel"]
specific_hosts = df_bk[df_bk["host_name"].isin(hosts)]

host_count = specific_hosts["host_name"].value_counts()
for host, count in host_count.items():
    print(f'{host}: {count} listings')

print(f'There are {len(specific_hosts)} listings altogether')
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 
df_bk["adjusted_price"] = np.where(df_bk["neighbourhood"] == "Williamsburg", df_bk["price"] + 100, df_bk["price"])
df_bk
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.
private_rooms_count = df_bk["room_type"].value_counts()  # Provide a zip variable with 
private_rooms_percentage = df_bk["room_type"].value_counts(normalize=True) * 100
specific_room_types = ["Private room", "Shared room"]

for room_type in private_rooms_percentage.index:
    if room_type in specific_room_types:
        print(f'{room_type}: {private_rooms_count[room_type]} rooms ({private_rooms_percentage[room_type]:.2f}%)')
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
listings_by_borough = listings.groupby("neighbourhood_group").size()
for borough, count in listings_by_borough.items():
    print(f'{borough}: {count} listings')
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
mean_price_per_borough = listings.groupby("neighbourhood_group")["price"].mean()
print(f'Mean Listing Price per Borough:')
for borough, mean_price in mean_price_per_borough.items():
    print(f'{borough}: ${mean_price:.2f}')
```

```python
# 2.5. Using `groupby`, find the mean price for each room_type.
mean_price_per_accommodation = listings.groupby("room_type")["price"].mean()
print(f'Mean Price Per Accommodation:')
for accommodation, price in mean_price_per_accommodation.items():
    print(f'{accommodation}: ${price:.2f}')
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
listings.groupby("neighbourhood_group")["price"].agg(["min", "max"])
```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.
mean_price_per_room_type_per_borough = listings.groupby(["neighbourhood_group", "room_type"])["price"].mean()
current_borough = None
for (borough, room_type), mean_price in mean_price_per_room_type_per_borough.items():
    if borough != current_borough:
        current_borough = borough
        print(f'\n{current_borough}')
        print('-' * len(current_borough))
    print(f'\t{room_type}: ${mean_price:.2f}')
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
listings.groupby(["neighbourhood_group", "room_type"])["price"].agg(["count", "min", "max", "mean", "median", "std"])
```

### (Extra Hard) Download the files for the past 5 years into a new folder and add them all into one data frame using pandas, then save that new file.
* Try to not hard code in the file names. We want to do this programmatically because what if we want to add new/more file names in the future.

```python
# I'll use the glob library which uses pattern matching to find pathnames
file_paths = glob.glob(f"../data/gl2020_24/gl*.txt") # Here we're telling the program to find all files that start with 'gl' in the gl2020_24 folder and return their paths
print(file_paths)

# file_paths is now an iterable list, We can create dataframes using each path in our list
dataframes = []
for file_path in file_paths:
    df = pd.read_csv(file_path, header=None)

    # A rudimentary dataframe is created, now we can use regex to extract the date to mark the orgin year
    file_name = file_path.split('/')[-1] # This will split the path on each directory name creating a list of each name and grab the file name which is the last element of the list
    match = re.search(r'gl(\d+)', file_name) # Then we do a regex search to grab all digits after we encounter the string 'gl'
    year = match.group(1) # This specifies the group to capture which in this case will be the year
    df["origin_year"] = year # Set the origin year column
    dataframes.append(df) # append the dataframe we are currently working on our accumulator
    print(f'Loaded {file_path}: {df.shape[0]} rows')

combined_df = pd.concat(dataframes, ignore_index=True) # Stack the dataframes together into a singular cohesive unit
combined_df.columns = column_names + ["origin_year"] # re-add the column names we generated in the previous cell and the origin year
combined_df = combined_df.sort_values('origin_year', ascending=False)
combined_df
```
